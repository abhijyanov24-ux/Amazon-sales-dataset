```python
from pandasql import sqldf
```

```python
import pandas as pd
data = pd.read_csv("amazon.csv")
query = "SELECT SUM(TotalAmount) AS Total_Sales FROM data"
result = sqldf(query, globals())

print(result)
```

```python
data = pd.read_csv("amazon.csv")
query = """
SELECT Category, SUM(TotalAmount) AS Total_Sales
FROM data
GROUP BY Category
ORDER BY Total_Sales DESC
"""
result = sqldf(query, globals())
print(result)
```

```python
query = """
SELECT City, SUM(TotalAmount) AS Total_Sales
FROM data
GROUP BY City
ORDER BY Total_Sales DESC
"""
result = sqldf(query, globals())
print(result)
```
